```python
precision = pd.read_csv("../../Pipeline/MultiClass/Order/precision_by_label.txt",sep=" = ",names=['family_id','precision'])
```

```python
family = pd.read_csv("../../Pipeline/MultiClass/all.vect.order.index", sep=" ",names = ['family','real_tax_id'])
```

```python
family
```

```python
new_family = precision.join(family)
```

```python
new_family
```

```python
group = pd.read_csv("../../Pipeline/MultiClass/order_group.list", sep=" ",names = ['group'],index_col=0)
```

```python
group
```

```python
new_precision = new_family.join(group, on="real_tax_id")
```

```python
precision_group = new_precision.loc[:,['precision','group']]
```

```python
precision_group
```

```python
precision_group.groupby(['group']).count()
```

```python
precision_group.groupby(['group']).describe()
```

```python
precision_group.hist(column='precision',by='group',bins=20,figsize=(10, 10))
plt.suptitle("precision",fontsize=16)
```

```python
recall = pd.read_csv("../../Pipeline/MultiClass/Order/recall_by_label.txt",sep=" = ",names=['family_id','recall'])
new_family_recall = recall.join(family)
new_recall = new_family_recall.join(group, on="real_tax_id")
recall_group = new_recall.loc[:,['recall','group']]
```

```python
recall_group.hist(column='recall',by='group',bins=20,figsize=(10, 10))
plt.suptitle("recall",fontsize=16)
```

```python
f1 = pd.read_csv("../../Pipeline/MultiClass/Order/f1_by_label.txt",sep=" = ",names=['family_id','f1'])
new_family_f1 = f1.join(family)
new_f1 = new_family_f1.join(group, on="real_tax_id")
f1_group = new_f1.loc[:,['f1','group']]
```

```python
f1_group.hist(column='f1',by='group',bins=20,figsize=(10, 10))
plt.suptitle("f1",fontsize=16)
```
