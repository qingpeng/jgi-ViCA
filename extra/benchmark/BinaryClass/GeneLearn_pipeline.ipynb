```python
def load_PRC_data(file_in,spark):
    # spark is output from spark pipeline or not(old dato)
    file_in_obj = open(file_in,'r')
    x = []
    y = []
    for line in file_in_obj:
        line = line.rstrip()
        if spark == "Y":
            line = line[1:-1]
            fields = line.split(',')

        else:
            fields = line.split()
        x.append(fields[0])
        y.append(fields[1])
    return x,y
```
