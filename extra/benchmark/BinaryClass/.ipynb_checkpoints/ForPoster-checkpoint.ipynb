```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def load_PRC_data(file_in,spark):
    # spark is output from spark pipeline or not(old dato)
    file_in_obj = open(file_in,'r')
    x = []
    y = []
    for line in file_in_obj:
        line = line.rstrip()
        if spark == "Y":
            line = line[1:-1]
            fields = line.split(',')

        else:
            fields = line.split()
        x.append(fields[0])
        y.append(fields[1])
    return x,y
```

Performance using different features (4mer, 5mer, codon, combined)
====


fasd

```python
# 5k segment training/testing, with codon + 5mer
x_5k_5mer_codon,y_5k_5mer_codon = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/5k_5mer_PRC.txt/all_logistic_PRC.txt","Y")
# 5k segment training/testing, with codon + 4mer
x_5k_4mer_codon,y_5k_4mer_codon = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/5k5k_PRC.txt/all.txt","Y")
# ~2000 RDRP contigs, ~10000 randomly selected segments from non-virus
x_RDRP,y_RDRP = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/RDRP_PRC.txt/all.txt","Y")
# only use 5 mer frequency
x_5k_5meronly,y_5k_5meronly = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/5k_5mer_only_PRC.txt/all.txt","Y")
# only use 4 mer frequency
x_5k_4meronly,y_5k_4meronly = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/5k_4mer_only_PRC.txt/all.txt","Y")
# only use codon/metagenemarkS frequency
x_5k_codon_only,y_5k_codon_only = load_PRC_data("/global/projectb/scratch/qpzhang/Full_Training/5k_codon_only_PRC.txt/all.txt","Y")
```

```python
plt.figure(figsize=(8, 6), dpi=80)


plt.plot(x_5k_5mer_codon,y_5k_5mer_codon,label = '5k-5mer-with-codon, auPRC:0.898')
plt.plot(x_5k_4mer_codon,y_5k_4mer_codon,label = '5k-4mer-with-codon, auPRC:0.884')

plt.plot(x_5k_5meronly,y_5k_5meronly,label = '5k-5mer-only, auPRC:0.793')
plt.plot(x_5k_4meronly,y_5k_4meronly,label = '5k-4mer-only, auPRC:0.729')
plt.plot(x_5k_codon_only,y_5k_codon_only,label = '5k-codon-only, auPRC:0.858')


plt.plot(x_RDRP,y_RDRP,label = 'RDRP_5kmodel-4mer')
plt.legend(loc='lower left')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('logistic, segment on segment, all data')
```

RDRP data
====

```python
# this file contains the predicted score and labels for each segments/contigs
file_RDRP = open("/global/projectb/scratch/qpzhang/Full_Training/RDRP_predictionAndLabels.txt/all.txt",'r')
```

```python
L2A = []
L2B = []
for line in file_RDRP:
    line = line.rstrip()
    line = line[1:-1]
    if line.split(',')[1] == '0.0':
        L2A.append(float(line.split(',')[0]))
    else:
        L2B.append(float(line.split(',')[0]))
```

```python
# example data

num_bins = 100
# the histogram of the data
n, bins, patches = plt.hist(L2A, num_bins, normed=1, facecolor='green', alpha=0.5, label='non-virus')
n, bins, patches = plt.hist(L2B, num_bins, normed=1, facecolor='red', alpha=0.5, label = 'virus')

plt.xlabel('Score')
plt.ylabel('Probability')
plt.title('Score,logistic classifier,RDRP(>5k length) and artificial negative 5K segments')
plt.legend()
plt.ylim(0,100)
# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.show()
```
